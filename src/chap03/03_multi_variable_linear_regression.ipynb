```python
import tensorflow as tf
import numpy as np
```

```python
x_data = [
          [73.,80.,75.],
          [93.,88.,93.],
          [89.,91.,90.,],
          [96.,98.,100.,],
          [73.,66.,70.]
        ]

y_data = [
          [152.],
          [185.],
          [180.],
          [196.],
          [142.]
        ]
```

```python
tf.model = tf.keras.Sequential() #모델 모형 갖추기 

tf.model.add(tf.keras.layers.Dense(input_dim=3, units=1)) # 신경세포 자료형 Dense 피쳐 3입력 1출력
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate = 1e-5))

tf.model.summary()
```

```python
loss=tf.model.fit(x_data, y_data, epochs=100)
```
