```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('../data/data-01-test-score.csv', delimiter=',', dtype=np.float32) #../ 상위 폴더 접근
x_data = xy[:, 0:-1] # 행은 전체읽어오기, 열은 마지막 열(정답데이터) 제외
y_data = xy[:, [-1]]
```

```python
print(x_data,'\nx_data shape :', x_data.shape)
print(y_data,'\ny_data shape :', y_data.shape)
```

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim =3, units=1, activation='linear'))
```

```python
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer = tf.keras.optimizers.SGD(learning_rate=1e-5))
```

```python
tf.model.fit(x_data, y_data, epochs=2000)
```

```python
print(tf.model.predict([[100., 70., 98,]]))
```
