```python
# 텐서플로우 설치확인
#conda install tensorflow

import tensorflow as tf
import numpy as np

print(tf.__version__)
```

```python
# y = wx + b : w = -1, b = 1

x_train = [1, 2, 3, 4] #1차원
y_train = [0, -1, -2, -3]  #1차원
```

```python
tf.model = tf.keras.Sequential() # 케라스 모델의 정의 되는 자료형Sequential, 모델의 틀
```

```python
# input_dim : input shape, units : output shape
tf.model.add(tf.keras.layers.Dense(input_dim=1, units=1))
```

```python
# SGD (Stochastic Gradient Descendent) :  기울기 구하기, lr(learning_rate) 
sgd = tf.keras.optimizers.SGD(lr=0.1)
```

```python
# mse(mean_squared_error), 1/m *sig(y` - y) ^ 2
tf.model.compile(loss ='mse', optimizer = sgd)
```

```python
weights = tf.model.layers[0].get_weights()
w = weights[0][0][0]
print('initial w is : ' + str(w))
```

```python
tf.model.summary()
```

```python
tf.model.fit(x_train, y_train, epochs=200) #훈련
```

```python
tf.model.predict(np.array([5])) #실행 
```

```python
# 학습을 통해 구한 w와 b의값
tf.model.layers[0].get_weights()
```

# 신경망 구현 순서
1. Sequential 모형 클래스 객체 생성
2. add 메서드로 레이어 추가.
  - 입력단부터 순차적으로 추가한다.
  - 레이어는 출력 뉴런 갯수를 첫번째 인수로 받는다.
  - 최초의 레이어는 input_dim 인수로 입력 크기를 설정해야 한다.
  - activation 인수로 활성화 함수 설정
3. compile 메서드로 모형 완성.
  - loss인수로 비용함수 설정
  - optimizer 인수로 최적화 알고리즘 설정
  - metrics 인수로 트레이닝 단계에서 기록할 성능 기준 설정
4. fit 메서드로 트레이닝
  - epochs로 에포크(epoch) 횟수 설정
