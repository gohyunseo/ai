```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = wx + b : w = -1, b = 1

x_train = [1, 2, 3, 4] #1차원
y_train = [0, -1, -2, -3]  #1차원

tf.model = tf.keras.Sequential() # 케라스 모델의 정의 되는 자료형Sequential, 모델의 틀

# input_dim : input shape, units : output shape
tf.model.add(tf.keras.layers.Dense(input_dim=1, units=1))

# SGD (Stochastic Gradient Descendent) :  기울기 구하기, lr(learning_rate) 
sgd = tf.keras.optimizers.SGD(lr=0.1)

# mse(mean_squared_error), 1/m *sig(y` - y) ^ 2
tf.model.compile(loss ='mse', optimizer = sgd)
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True' #라이브러리 충돌 방지

loss = tf.model.fit(x_train, y_train, epochs=200) #훈련
y_predict = tf.model.predict(np.array([5])) #실행 
y_predict
```

```python
# loss값의 시각화
plt.plot(loss.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc = 'upper left')
plt.show()
```
