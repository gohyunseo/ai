```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 단순한 그래프 그리기

# 데이터 준비
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

```python
# 그래프 그리기

plt.plot(x, y1, label = 'sin') #파란선
plt.plot(x, y2, linestyle = '--', label ='cos') #주황선
plt.xlabel('x') #x축 이름
plt.ylabel('y') #y축 이름
plt.title('sin & cos') #그래프 제목
plt.legend() #범례
plt.show() #출력
```

```python
#탄젠트
plt.plot(x, y3)
```

```python
#np.random.seed(1234) #난수 고정
x = np.arange(10)
y = np.random.rand(10) #0 ~ 1 사이의 균일한 분포

plt.plot(x,y) #꺽은선 그래프를 등록

plt.show()
```

```python
# 3차 함수 f(x) = (x-2) * x *(x+2)
```

```python
def f(x):    #함수 정의
    return (x-2) * x * (x+2)

print(f(0))
print(f(2))
print(f(-2))
```

```python
# x 값에 대해 ndarray 배열이며 각각에 대응 한 f를 한꺼번에 ndarray로 돌려줌.
print(f(np.array([1, 2, 3])))
```

```python
x = np.arange(-3, 3.5, 0.5)
y =  f(np.array([1, 2, 3]))
plt.plot(x, y)
plt.show
```

```python
# 그래프를 장식
def f2(x, w):
    return (x-w) * x * (x+2) #함수 정의
```

```python
# x를 정의
x = np.linspace(-3, 3, 100) #x를 100분할
```

```python
plt.plot(x,f2(x,2), color='black', label ='f2(x,2)')
plt.plot(x,f2(x,1), color='blue', label ='f2(x,1)')
plt.legend(loc = 'upper left')
plt.ylim(-15, 15) #y축 범위
plt.title('f2(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True) #그리드 (눈금)
plt.show()
```

```python
# 그래프 여러개 보여주기

plt.figure(figsize=(10, 3)) # 전체 영역의 크기를 지정
plt.subplots_adjust(wspace= 0.5, hspace= 0.5) #그래프의 간격 지정

for i in range(6):        #0~5까지 반복
    plt.subplot(2, 3, i+1) #그래프의 위치를 지정
    plt.title(i+1)
    plt.plot(x, f2(x, i))
    plt.ylim(-20, 20)
    plt.grid(True)
    
plt.show()
```
