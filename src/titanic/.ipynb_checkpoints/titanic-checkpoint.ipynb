```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(train.shape[0])
print(train.shape[1])
print(test.shape[0])
print(test.shape[1])
print(train.head())
```

```python
sns.lmplot(x = 'Survived', y = 'Age',data = train, fit_reg = False, # x, y, , 
               scatter_kws = {"s":100}, #  
               markers = ['o','x'],
               hue ='Sex')
```

```python
sns.countplot(x = 'Sex',  data = train, hue = 'Survived')
```
