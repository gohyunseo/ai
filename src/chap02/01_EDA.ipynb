# 데이터 수집

```python
#데이터를 수집 
df = pd.read_csv('data/csv/basketball_stat.csv')
df
```

# 데이터시각화
- 데이터 특징을 바탕으로 한 공간에 시각화 함으로써, 우리는 머신러닝 학습에 필요한 특징과 불필요한 특징을 쉽게 구분지을 수 있다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 스틸, 2점슛 데이터 시각화
sns.lmplot(x ='STL', y = '2P', data = df, fit_reg = False, # x축, y축, 데이터, 라인없음
               scatter_kws = {"s":150}, #좌표상의 점의 크기
               markers = ['o','x'],
               hue ='Pos') #예측값

#title
plt.title('STL and 2P in 2d place')
```

```python
# 어시스트, 2점슛 데이터 시각화
sns.lmplot(x ='AST', y = '2P', data = df, fit_reg = False, # x축, y축, 데이터, 라인없음
               scatter_kws = {"s":150}, #좌표상의 점의 크기
               markers = ['o','x'],
               hue ='Pos') #예측값

#title
plt.title('AST and 2P in 2d place')
```

```python
# 블록킹, 3점슛 데이터 시각화
sns.lmplot(x ='BLK', y = '3P', data = df, fit_reg = False, # x축, y축, 데이터, 라인없음
               scatter_kws = {"s":150}, #좌표상의 점의 크기
               markers = ['o','x'],
               hue ='Pos') #예측값

#title
plt.title('BLK and 3P in 2d place')
```

```python
# 리바운드, 3점슛 데이터 시각화
sns.lmplot(x ='TRB', y = '3P', data = df, fit_reg = False, # x축, y축, 데이터, 라인없음
               scatter_kws = {"s":150}, #좌표상의 점의 크기
               markers = ['o','x'],
               hue ='Pos') #예측값

#title
plt.title('TRB and 3P in 2d place')
```

# 데이터 전처리

```python
# 분별력이 없는 특징(feature)을 데이터에서 제거, 데이터 다듬기

df.drop(['2P','AST','STL'], axis = 1, inplace = True)
df.head()
```

# 데이터 나누기 (학습데이터, 테스트데이터)

```python
# scikit-learn library 설치
#!conda install scikit-learn
```

```python
# skilearn 의 train_test_split 데이터를 사용하면 코드한줄로 손쉽게 코드 나누기 가능
from sklearn.model_selection import train_test_split

# 다듬어진 데이터에서 20%를 테스트 데이터 분류.
train, test = train_test_split(df, test_size=0.2)
```

```python
# 학습데이터의 갯수
train.shape[0]
```

```python
# 테스트 데이터의 갯수,
test.shape[0]
```

```python
train.to_csv('data/csv/basketball_train.csv', index = None)
```

```python
test.to_csv('data/csv/basketball_test.csv', index = None)
```
